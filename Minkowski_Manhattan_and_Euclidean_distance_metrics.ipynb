# **M√©tricas de Dist√¢ncia: Manhattan, Euclidiana e Minkowski**

Em aprendizado de m√°quina, a escolha da m√©trica de dist√¢ncia adequada √© crucial para o sucesso de algoritmos como k-NN e k-means. As m√©tricas de Minkowski, Euclidiana e Manhattan s√£o algumas das mais utilizadas.

**M√©trica de Manhattan:** Tamb√©m conhecida como dist√¢ncia do taxista, calcula a dist√¢ncia como a soma das diferen√ßas absolutas entre as coordenadas dos pontos. Corresponde √† m√©trica de Minkowski com `p = 1`.

**M√©trica Euclidiana:** √â a m√©trica mais comum, que calcula a dist√¢ncia em linha reta entre dois pontos. Corresponde √† m√©trica de Minkowski com `p = 2`.

**M√©trica de Minkowski:** √â uma fam√≠lia de m√©tricas que inclui as m√©tricas Euclidiana e Manhattan como casos especiais. A m√©trica de Minkowski de ordem `p` √© calculada como a raiz `p`-√©sima da soma das diferen√ßas absolutas elevadas √† pot√™ncia `p` entre as coordenadas dos pontos.

A escolha da m√©trica depende da natureza dos dados e do problema em quest√£o. A m√©trica Euclidiana √© frequentemente usada quando a dist√¢ncia "natural" entre os pontos √© relevante, enquanto a m√©trica de Manhattan pode ser mais apropriada em casos onde o movimento √© restrito a eixos perpendiculares, como em uma grade. A m√©trica de Minkowski oferece flexibilidade para ajustar a sensibilidade da m√©trica √†s diferen√ßas entre as dimens√µes.


## 1) **M√©trica de Manhattan** (ou norma L1)

conhecida como "dist√¢ncia do taxista", calcula a dist√¢ncia entre dois pontos como a soma das diferen√ßas absolutas de suas coordenadas em cada dimens√£o. Imagine um trajeto em uma cidade com ruas em formato de grade: a dist√¢ncia de Manhattan corresponde ao n√∫mero de quarteir√µes percorridos horizontalmente e verticalmente para chegar ao destino. Essa m√©trica √© particularmente √∫til em cen√°rios onde o movimento se restringe a eixos perpendiculares, como em malhas urbanas, jogos de tabuleiro ou an√°lise de imagens.

Exemplos de aplica√ß√µes:

* Rob√≥tica: Planejamento de rotas para rob√¥s em ambientes com obst√°culos, onde os movimentos s√£o limitados a dire√ß√µes ortogonais.
* Log√≠stica: C√°lculo de dist√¢ncias em cidades para otimizar rotas de entrega.
* Gen√¥mica: Compara√ß√£o de sequ√™ncias de DNA, onde as diferen√ßas entre as bases nitrogenadas s√£o medidas em termos de substitui√ß√µes, inser√ß√µes e dele√ß√µes.

## 2) **M√©trica Euclidiana** (ou norma L2)

√â definida pela dist√¢ncia entre dois pontos como o comprimento da linha reta que os conecta. √â a m√©trica mais intuitiva e amplamente utilizada em diversas √°reas, incluindo algoritmos de aprendizado de m√°quina como o k-means.

O k-means visa agrupar pontos de dados em clusters com base em sua similaridade. A m√©trica euclidiana √© empregada para calcular a dist√¢ncia entre cada ponto e os centr√≥ides dos clusters, permitindo que os pontos sejam atribu√≠dos ao cluster mais pr√≥ximo. Esse processo iterativo de atribui√ß√£o e atualiza√ß√£o dos centr√≥ides continua at√© que a configura√ß√£o dos clusters se estabilize, minimizando a soma das dist√¢ncias quadradas entre os pontos e seus respectivos centr√≥ides.

O objetivo do k-means √© minimizar a soma das dist√¢ncias quadradas entre cada ponto e o centro do cluster ao qual ele pertence, particionando os dados em ùëò clusters, agrupando pontos semelhantes e separando pontos distintos. A m√©trica euclidiana calcula a dist√¢ncia em linha reta entre dois pontos em um espa√ßo multidimensional, sendo  usada para determinar a proximidade entre os pontos e, assim, atribu√≠-los aos clusters.

Exemplos de aplica√ß√µes:
* Reconhecimento de padr√µes, a m√©trica euclidiana √© aplicada para medir a similaridade entre diferentes padr√µes num√©ricos. Por exemplo, em um sistema de reconhecimento facial, as caracter√≠sticas faciais podem ser representadas como vetores num√©ricos, e a dist√¢ncia euclidiana entre esses vetores pode ser usada para comparar rostos e determinar se eles pertencem √† mesma pessoa.

* Vis√£o computacional, a m√©trica euclidiana √© frequentemente usada para avaliar dist√¢ncias entre pixels em imagens. Por exemplo, em algoritmos de segmenta√ß√£o de imagens, a dist√¢ncia euclidiana entre as cores de pixels adjacentes pode ser usada para determinar se eles pertencem ao mesmo objeto ou a diferentes regi√µes da imagem.

## 3) **M√©trica de Minkowski**

A m√©trica de Minkowski √© uma generaliza√ß√£o que inclui tanto a dist√¢ncia Euclidiana (ùëù=2) quanto a dist√¢ncia de de Manhattan (ùëù=1) como casos especiais - ermite ajustar o par√¢metro "p" para controlar a influ√™ncia de cada dimens√£o no c√°lculo da dist√¢ncia. A m√©trica Euclidiana representa a dist√¢ncia direta ("linha reta") entre dois pontos no espa√ßo. Por sua vez, a m√©trica de Manhattan (tamb√©m chamada de dist√¢ncia do "taxista"), mede a dist√¢ncia entre dois pontos somando os deslocamentos absolutos em cada dimens√£o.

**Par√¢metros**:
* ùëù, ùëû: Vetores de pontos no espa√ßo.
* ùëõ: N√∫mero de dimens√µes.
* ùëù: Ordem da m√©trica.

**Casos Especiais**:
* ùëù=1: M√©trica de Manhattan (ùêø1), adequado para dados esparsos ou onde os eixos s√£o independentes.
* ùëù=2: M√©trica Euclidiana (ùêø2), √∫til para dados cont√≠nuos e esf√©ricos.
* ùëù‚Üí‚àû: M√©trica do Chebyshev (ùêø‚àû) (max‚à£ùëùùëñ‚àíùëûùëñ‚à£); com ùëù>2 s√£o real√ßadas grandes diferen√ßas, sendo √∫til para identificar outliers.

√â usada em algoritmos de classifica√ß√£o e regress√£o em *machine learn* como k-Nearest Neighbors (k-NN), por exemplo, e em problemas de otimiza√ß√£o para avalia√ß√£o de diferen√ßas em m√©tricas personalizadas.

No k-NN, a m√©trica de Minkowski √© usada para calcular dist√¢ncias entre o ponto de teste e os exemplos do conjunto de treinamento. Dado um ponto ùë•, √© calculada a dist√¢ncia entre ùë• e todos os exemplos no conjunto de treinamento, em seguida, o algoritmo ordena os exemplos por proximidade e seleciona os
ùëò-"vizinhos mais pr√≥ximos". A classe mais comum entre os ùëò-vizinhos √© atribu√≠da ao ponto ùë•.

Exemplos de aplica√ß√µes interessantes:
* An√°lise de dados, para descobrir padr√µes ocultos em grandes conjuntos de dados.
* Segmenta√ß√£o de Clientes, com base em comportamentos semelhantes para campanhas de marketing direcionadas.
* Bioinform√°tica, para classifica√ß√£o de prote√≠nas com base em suas estruturas tridimensionais.
Finan√ßas: Detec√ß√£o de fraudes em transa√ß√µes financeiras, identificando padr√µes incomuns.
* Processamento e Compress√£o de Imagens: para clusteriza√ß√£o de regi√µes em imagens com base em cores ou texturas, ou para reduzir cores de uma imagem, agrupando pixels com valores semelhantes.
* Reconhecimento de d√≠gitos manuscritos (MNIST), em que o k-NN classifica imagens de d√≠gitos, baseando-se nas dist√¢ncias entre vetores de pixels.

```python
import pandas as pd
import pylab
from tabulate import tabulate
```

```python
def minkowskiDist(v1, v2, p):
#def minkowskiDist(v1: list[float], v2: list[float], p: int) -> float:
    """
    Calcula a dist√¢ncia de Minkowski entre dois vetores.

    Par√¢metros:
        v1: lista de n√∫meros (vetor 1).
        v2: lista de n√∫meros (vetor 2).
        p: ordem da m√©trica de Minkowski.

    Retorna:
        A dist√¢ncia de Minkowski de ordem p entre os vetores v1 e v2.
    """
    if len(v1) != len(v2):
        raise ValueError("Os vetores devem ter o mesmo comprimento.")

    dist = 0.0
    for i in range(len(v1)):
        dist += abs(v1[i] - v2[i])**p
    return dist**(1/p)
```

```python
class Animal(object):
    """
    Representa um animal com nome e caracter√≠sticas.

    Atributos:
        name (str): Nome do animal.
        features (list): Vetor de caracter√≠sticas do animal.

    M√©todos:
        __init__(self, name, features): Inicializa o animal com nome e caracter√≠sticas.
        getName(self): Retorna o nome do animal.
        getFeatures(self): Retorna o vetor de caracter√≠sticas do animal.
        distance(self, other): Calcula a dist√¢ncia euclidiana entre este animal e outro.
        __str__(self): Retorna o nome do animal em formato de string.
    """

    def __init__(self, name, features):
        """
        Inicializa o objeto Animal.

        Par√¢metros:
            name: Nome do animal (string).
            features: Lista de caracter√≠sticas do animal (n√∫meros).
        """
        self.name = name
        self.features = pylab.array(features)

    def getName(self):
        """Retorna o nome do animal."""
        return self.name

    def getFeatures(self):
        """Retorna o vetor de caracter√≠sticas do animal."""
        return self.features

    def distance(self, other, p=2):
        """
        Calcula a dist√¢ncia euclidiana entre este animal e outro.

        Par√¢metros:
            other: Outro objeto Animal.

        Retorna:
            Dist√¢ncia euclidiana (float).
        """
        return minkowskiDist(self.getFeatures(), other.getFeatures(), 2)

    def __str__(self):
        """Retorna o nome do animal como string."""
        return self.name

    def __hash__(self):
        """Define a fun√ß√£o hash para objetos Animal com base no nome e nas caracter√≠sticas."""
        return hash((self.name, tuple(self.features))) # Convert features to tuple for hashability

    def __eq__(self, other):
        """Compara dois animais com base em seus nomes e caracter√≠sticas."""
        return self.name == other.name and all(self.features == other.features)

# Nome do aninal e vetor de caracter√≠sticas: posi√ß√£o de ovos, escamas, venenoso, sangue frio, n√∫mero de pernas, r√©ptil
cobra = Animal('Cobra', [1, 1, 0, 1, 0, 1]) # Nem todas as cobras s√£o venenosas
cascavel = Animal('Cascavel', [1, 1, 1, 1, 0, 1])
galinha = Animal('Galinha', [1, 1, 0, 0, 2, 0])
guppyFish = Animal('Guppy', [0, 1, 0, 0, 0, 0])
jacare = Animal('Jacar√©', [1, 1, 0, 1, 4, 1])
jiboia = Animal('Jib√≥ia', [0, 1, 0, 1, 0, 1])
python = Animal('Python', [1, 1, 0, 1, 0, 1])
salmao = Animal('Salm√£o', [1, 1, 0, 1, 0, 0])
sapoDardo = Animal('Sapo-dardo', [1, 0, 1, 0, 4, 0])
zebra = Animal('Zebra', [0, 0, 0, 0, 4, 0])

animals = [cobra, cascavel, galinha, guppyFish, jacare, jiboia, salmao, sapoDardo, python, zebra]
```

```python
def compareAnimals(animals, precision):
    """
    Constr√≥i uma tabela de dist√¢ncias euclidianas entre pares de animais.

    Par√¢metros:
        animals: Lista de objetos Animal.
        precision: Precis√£o do n√∫mero decimal nas dist√¢ncias (int >= 0).

    Retorna:
        Gera uma tabela visual das dist√¢ncias euclidianas entre os animais e salva como imagem.
    """
    if not all(isinstance(a, Animal) for a in animals):
        raise TypeError("Todos os elementos da lista devem ser inst√¢ncias da classe Animal.")

    # Inicializa r√≥tulos para as colunas e linhas
    columnLabels = []
    for a in animals:
        columnLabels.append(a.getName())
    rowLabels = columnLabels[:]
    tableVals = []

    # Calcula dist√¢ncias entre pares de animais
    for a1 in animals:
        row = []
        for a2 in animals:
            if a1 == a2:
                row.append('--')  # Dist√¢ncia de um animal para ele mesmo
            else:
                distance = a1.distance(a2)
                row.append(str(round(distance, precision)))
        tableVals.append(row)

    # Gera a tabela usando PyLab
    table = pylab.table(
        rowLabels=rowLabels,
        colLabels=columnLabels,
        cellText=tableVals,
        cellLoc='center',
        loc='center',
        colWidths=[0.2] * len(animals)
    )
    table.scale(1, 2.5)
    pylab.savefig('distances')
    return table

distances = {}
for i, a1 in enumerate(animals):
    for j, a2 in enumerate(animals):
        if i < j:
            distances[(a1, a2)] = a1.distance(a2)

data = []
for i in range(len(animals)):
    for j in range(len(animals)):
        if i < j:
            a = animals[i]
            b = animals[j]
            dist = round(a.distance(b), 3)
            print('Dist√¢ncia entre', a, 'e', b, '√©', dist)
```

```python
# Criar um DataFrame com as dist√¢ncias
data = []
for i in range(len(animals)):
    for j in range(len(animals)):
        if i <= j:  # Inclui diagonal com dist√¢ncia 0
            a = animals[i]
            b = animals[j]
            dist = round(a.distance(b), 3) if i != j else 0
            data.append([a.getName(), b.getName(), dist])

# Criar o DataFrame
df = pd.DataFrame(data, columns=['Animais', 'Cross-tab', 'Dist√¢ncia'], )

# Criar a tabela cruzada com os animais ordenados
unique_animals = [a.getName() for a in animals]
pivot_table = df.pivot(index='Animais', columns='Cross-tab', values='Dist√¢ncia')
pivot_table = pivot_table.reindex(index=unique_animals, columns=unique_animals, fill_value=0)

# Exibir a tabela
#print(pivot_table)
#print(pivot_table.to_string(index=True, header=True))
print(tabulate(pivot_table, headers='keys', tablefmt='psql'))
```

A tabela cruzada exibe as dist√¢ncias entre diferentes animais, est√° sendo usada para agrupar animais similares (ex.: clusteriza√ß√£o) que foram calculadas com base em seus vetores de caracter√≠sticas. Tamb√©m pode servr para classificar novos exemplos com base na proximidade a animais existentes. A dist√¢ncia utilizada aqui √© Euclidiana (ùëù=2), que mede a reta mais curta entre dois pontos no espa√ßo de caracter√≠sticas.
